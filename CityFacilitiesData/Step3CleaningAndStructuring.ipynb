```python
import pandas as pd
import re

# Standardizing street type abbreviations
street_abbreviations = {
    " ST": " Street",
    " RD": " Road",
    " AVE": " Avenue",
    " BLVD": " Boulevard",
    " DR": " Drive",
    " LN": " Lane",
    " PIKE": " Pike",
    " PKWY": " Parkway",
    " CIR": " Circle",
    " CT": " Court",
    " HWY": " Highway"
}

# Function to clean and standardize addresses
def standardize_address(location):
    if pd.isna(location):  # Handle missing values
        return location

    # Ensure location is a string
    location = str(location).strip().title()

    # Replace common abbreviations with full names
    for abbr, full in street_abbreviations.items():
        location = re.sub(rf"{abbr}\b", full, location, flags=re.IGNORECASE)

    # Ensure consistent formatting with Bloomington, IN and zip code
    if "Bloomington" not in location:
        location += ", Bloomington, IN"

    return location

# Apply function to standardize Location column
city_fac_bloomington_full["Location"] = city_fac_bloomington_full["Location"].apply(standardize_address)
```

```python
from collections import defaultdict

# Helper to get initials
def get_initials(text):
    return ''.join([word[0] for word in text.split() if word[0].isalnum()]).upper()

# Dictionary to track counts for each (Group Initials + Type Initials) pair
counter = defaultdict(int)

# Function to generate ID
def generate_facility_id(row):
    group_initials = get_initials(row['Facility Group'])
    type_initials = get_initials(row['Facility Type'])
    key = f"{group_initials}-{type_initials}"
    counter[key] += 1
    number = str(counter[key]).zfill(3)  # zero-pad to 3 digits
    return f"{key}-{number}"

# Apply to DataFrame
city_fac_bloomington_full['Facility ID'] = city_fac_bloomington_full.apply(generate_facility_id, axis=1)

# Move Facility_ID to the first column
cols = city_fac_bloomington_full.columns.tolist()
cols = ['Facility ID'] + [col for col in cols if col != 'Facility ID']
city_fac_bloomington_full = city_fac_bloomington_full[cols]
```
