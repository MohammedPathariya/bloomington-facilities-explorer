```python
import pandas as pd
import re

# Standardizing street type abbreviations
street_abbreviations = {
    " ST": " Street",
    " RD": " Road",
    " AVE": " Avenue",
    " BLVD": " Boulevard",
    " DR": " Drive",
    " LN": " Lane",
    " PIKE": " Pike",
    " PKWY": " Parkway",
    " CIR": " Circle",
    " CT": " Court",
    " HWY": " Highway"
}

# Function to clean and standardize addresses
def standardize_address(location):
    if pd.isna(location):  # Handle missing values
        return location

    # Ensure location is a string
    location = str(location).strip().title()

    # Replace common abbreviations with full names
    for abbr, full in street_abbreviations.items():
        location = re.sub(rf"{abbr}\b", full, location, flags=re.IGNORECASE)

    # Ensure consistent formatting with Bloomington, IN and zip code
    if "Bloomington" not in location:
        location += ", Bloomington, IN"

    return location

# Apply function to standardize Location column
city_fac_bloomington["Location"] = city_fac_bloomington["Location"].apply(standardize_address)

# Save the cleaned dataset
upd_city_fac_bloomington.to_csv("city_fac_bloomington_standardized.csv", index=False)

# Display a preview of updated locations
print(upd_city_fac_bloomington[["Facility", "Location"]].head(20))
```
