```python
import requests
import time

def get_coordinates(address):
    url = "https://nominatim.openstreetmap.org/search"
    params = {
        "q": address,
        "format": "json",
        "limit": 1
    }
    headers = {
        "User-Agent": "Mozilla/5.0 (compatible; MyApp/1.0; +http://yourwebsite.com)"
    }
    try:
        response = requests.get(url, params=params, headers=headers)
        data = response.json()
        if data:
            return data[0]["lat"], data[0]["lon"]
        else:
            return None, None
    except Exception as e:
        print(f"Error fetching {address}: {e}")
        return None, None

# Example list of addresses
addresses = ["7470 S Shields Ridge Road, Bloomington, IN", "405 W Patterson Drive, Bloomington, IN", "108 E Kirkwood Avenue, Bloomington, IN", "407 W Patterson Drive, Bloomington, IN", "412 E 6Th Street, Bloomington, IN", "102 E Kirkwood Avenue, Bloomington, IN", "350 W 11Th Street, Bloomington, IN", "614 E 2Nd Street, Bloomington, IN", "301 E 3Rd Street, Bloomington, IN", "413 E 4Th Street, Bloomington, IN", "430 E 4Th Street, Bloomington, IN", "408 E 4Th Street, Bloomington, IN", "100 E Kirkwood Avenue, Bloomington, IN", "300 E 3Rd Street, Bloomington, IN", "900 E 7Th Street, Bloomington, IN", "223 W 6Th Street, Bloomington, IN", "106 N Walnut Street, Bloomington, IN", "125 N College Avenue, Bloomington, IN", "409 E 4Th Street, Bloomington, IN", "430 E Kirkwood Avenue Unit 6, Bloomington, IN", "4600 State Highway 446, Bloomington, IN", "2620 E 3Rd Street, Bloomington, IN", "250 S Washington Street, Bloomington, IN", "2500 W 3Rd Street, Bloomington, IN", "108 E Kirkwood Avenue, Bloomington, IN", "900 E 7Th Street, Bloomington, IN", "316 E 4Th Street, Bloomington, IN", "2632 E 3Rd Street, Bloomington, IN", "229 S Grant Street Suite 1, Bloomington, IN", "108 E 6Th Street, Bloomington, IN", "415 E 4Th Street, Bloomington, IN", "200 Daniels Way, Bloomington, IN", "2088 Liberty Drive, Bloomington, IN", "444 S Patterson Drive, Bloomington, IN", "1650 W Bloomfield Road, Bloomington, IN", "2300 E 2Nd Street, Bloomington, IN", "1315 East 10Th Street, Bloomington, IN", "500 W 7Th Street, Bloomington, IN", "952 S Kirby Road, Bloomington, IN", "1000 W 15Th Street, Bloomington, IN", "Ellettsville, In, Bloomington, IN", "200 Daniels Way, Bloomington, IN", "4001 E 3Rd Street Suite 1, Bloomington, IN", "4609 In-46, Bloomington, IN", "800 E Kirkwood Avenue, Bloomington, IN", "3230 S Walnut Street, Bloomington, IN", "349 S Walnut Street, Bloomington, IN", "7587 Rockport Road, Bloomington, IN", "1807 S Highland Avenue, Bloomington, IN", "2224 E 3Rd Street, Bloomington, IN", "300 W Clear Creek Drive, Bloomington, IN", "Ellettsville, In, Bloomington, IN", "4423 E Trailridge Road, Bloomington, IN", "417 N Elm Street, Bloomington, IN"]

# Fetch coordinates
for addr in addresses:
    lat, lon = get_coordinates(addr)
    print(f"{addr} => Latitude: {lat}, Longitude: {lon}")
    time.sleep(1)  # prevent hitting the rate limit
```
